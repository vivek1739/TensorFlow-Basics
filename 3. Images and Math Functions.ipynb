## Representing Image as a 3D tensor
- each pixel holds some value 
- in color image, pixel has RGB value ie. value of a particular color. In color image each pixel requires 3 values  to represent info within it. A value of R , G, B 
- value is 0-255
    - example : 0,255,0 ie. one pixel three values
- the numbr of values required to represent an image is the numbr of channel the image has
    - color image has three channels
    
*Grey Scale Images*
- contains shades of grey
- here every pixel contains exactly one value
- pixel reprensents intensity and range is 0.0-1.0
- we can represent in higher range but 0 -1 is common
- 1 value to represent a pixel .'. 1 channel

Note : Greyscale image (6,6,1) , color image (6,6,3)
- Tensors are nothing but n-d arrays .'. these images can be represented as tensors easily and 1 image is a 3d tensor

```python
import tensorflow as tf
import matplotlib.image as mp_img
import matplotlib.pyplot as plot
import os

% matplotlib inline

filename = "./Seed.JPG"
# it holds the image as a numpy array
image = mp_img.imread(filename)

print("Image shape: ",image.shape)
print("Image array",image)

plot.imshow(image)
plot.show()
```

```python
# hold the variable as a tensor
x = tf.Variable(image, name='x')

init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    # transpose will flip the width and height
    # normal is 0,1,2. We tranpose on 1st and 2nd axis and leave 3rd
        # as it is
    transpose = tf.transpose(x, perm=[1,0,2])
    
    #or we can use transpose meant only for images
    # transpose = tf.image.transpose_image(x)
    
    result = sess.run(transpose)
    
    print("Transposed image shape: ",result.shape)
    plot.imshow(result)
    plot.show()
    
```

***Resizing Images***


```python
import tensorflow as tf

from PIL import Image

original_image_list = ["./1.jpg","./2.jpg","./3.jpg","./4.jpg"]
# these are RGB images but black and white

# using queues construct in TF to read these files
# make a queue of file names with the given list
filename_queue = tf.train.string_input_producer(original_image_list)

# Read an entire image file
image_reader = tf.WholeFileReader()

with tf.Session() as sess:
    # NOTE: A session object in TF is multithreaded and we want 
        # to use multiple threads to read the image file
    # we can instantiate coordinator class, to make working 
        # with threads and queue easy
    coord = tf.train.Coordinator()
    # a coordinator abstracts all queue related threading activites 
    # ie. all threads must be able to stop together, exceptions need 
        # to be caught and reported, queues need to be closed properly
    # queues are a very convenient way to compute tensors asyncronously using 
        # multiple threads
    threads = tf.train.start_queue_runners(sess=sess, coord=coord)

    # storing all image tensors in a image list
    image_list = [];
    # using for loop to iterate number of images, so that we 
        # can dequeue from the queue
    for i in range(len(original_image_list)):
        # read a whole file from the head of queue
        # the first returning value in the tuple is the filename
            # so we will ignore
        _, image_file = image_reader.read(filename_queue)

        # decode the image as a jpeg file, this will turn into a tensor
        image = tf.image.decode_jpeg(image_file)

        # get a tensor of resized images
        image = tf.image.resize_images(image, [224,224])
        image.set_shape((224,224,3))

        # Get an image tensor and print its value
        image_array = sess.run(image)
        print(image_array.shape)
        
        # display image to screen
        Image.fromarray(image_array.astype('uint8'), 'RGB').show()

        # we want to convert multiple 3d images to one 4d
        # expand_dims adds a new dimension
        image_list.append(tf.expand_dims(image_array,0))

    # Finish of the filename queue coordinator
    coord.request_stop()
    coord.join(threads)

    index = 0

    # write image summary 
    summary_writer = tf.summary.FileWriter('./m4_example2',graph=sess.graph)

    for image_tensor in image_list:
        # summary string is summary representation of an image
        summary_str = sess.run(tf.summary.image("image-"+str(index),image_tensor))
        summary_writer.add_summary(summary_str)
        index+=1

    summary_writer.close()
```

### List of Images as 4D Tensors
- all images need to be of same size
- images with diff dimensions and different number of channels are not represented as a 4D tensor
- ex. (10,6,6,3) ; 10 is numbr of images

```python
import tensorflow as tf

from PIL import Image

original_image_list = ["./1.jpg","./2.jpg","./3.jpg","./4.jpg"]
# these are RGB images but black and white

# using queues construct in TF to read these files
# make a queue of file names with the given list
filename_queue = tf.train.string_input_producer(original_image_list)

# Read an entire image file
image_reader = tf.WholeFileReader()

with tf.Session() as sess:
    # NOTE: A session object in TF is multithreaded and we want 
        # to use multiple threads to read the image file
    # we can instantiate coordinator class, to make working 
        # with threads and queue easy
    coord = tf.train.Coordinator()
    # a coordinator abstracts all queue related threading activites 
    # ie. all threads must be able to stop together, exceptions need 
        # to be caught and reported, queues need to be closed properly
    # queues are a very convenient way to compute tensors asyncronously using 
        # multiple threads
    threads = tf.train.start_queue_runners(sess=sess, coord=coord)

    # storing all image tensors in a image list
    image_list = [];
    # using for loop to iterate number of images, so that we 
        # can dequeue from the queue
    for i in range(len(original_image_list)):
        # read a whole file from the head of queue
        # the first returning value in the tuple is the filename
            # so we will ignore
        _, image_file = image_reader.read(filename_queue)

        # decode the image as a jpeg file, this will turn into a tensor
        image = tf.image.decode_jpeg(image_file)

        # get a tensor of resized images
        image = tf.image.resize_images(image, [224,224])
        image.set_shape((224,224,3))
        
        # other image operations
        image = tf.image.flip_up_down(image)
        
        # central crop
        image = tf.image.central_crop(image,central_fraction=0.5)

        # Get an image tensor and print its value
        image_array = sess.run(image)
        print(image_array.shape)
        
        # display image to screen
        #Image.fromarray(image_array.astype('uint8'), 'RGB').show()
        
        # converts a numpy array of (224,224,3) to a tensor of shape (224,224,3)
        # generally stack is used to conver Rank R tensor list to one rank R+1 tensor
        # since here it is a numpy array so it just converts to tensor 
        image_tensor= tf.stack(image_array)

        print(image_tensor)
        image_list.append(image_tensor)

    # Finish of the filename queue coordinator
    coord.request_stop()
    coord.join(threads)
    
    # Now we can run stack to get 4d out of list of 3D
    images_tensor = tf.stack(image_list)
    print(images_tensor)

    summary_writer = tf.summary.FileWriter('./m4_example3', graph=sess.graph)

    # Having a 4D tensor will make writing the summary easier
    # by default it writes 3 images only
    summary_str = sess.run(tf.summary.image("images",images_tensor,max_outputs=4))
    summary_writer.add_summary(summary_str)
```
